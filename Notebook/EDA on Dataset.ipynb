## Remove unwanted columns

```python
drop_cols=['transaction_frequency_24h','avg_transaction_amount_7d']
df.drop(drop_cols,axis=1,inplace=True)
```

```python
df.drop('failed_transaction_count_24h',axis=1,inplace=True)
```

```python
## Feature Engineering
df['Date']=df['transaction_timestamp'].str.split('-').str[2]
df['Month']=df['transaction_timestamp'].str.split('-').str[1]
df['Year']=df['transaction_timestamp'].str.split('-').str[0]
```

```python
df.info()
```

```python
df.head()
```

```python
df['Date']=df['Date'].str.split(' ').str[0]
```

```python
df.head()
```

```python
df.drop('transaction_timestamp',axis=1,inplace=True)
```

```python
df['transaction_id'] = df['transaction_id'].str[1:]
df['customer_id'] = df['customer_id'].str[1:]
```

```python
df.info()
```

```python
df['transaction_id'] = df['transaction_id'].astype(int)
df['customer_id'] = df['customer_id'].astype(int)
df['Transaction_Date'] = df['Date'].astype(int)
df['Transaction_Month'] = df['Month'].astype(int)
df['Transaction_Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
drop_cols = ['Date','Month','Year']
df.drop(drop_cols,axis=1,inplace=True)
```
